```python
import pandas as pd
import numpy as np
```

```python
# Charge URLs 
file1_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
file2_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
file3_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(file1_url)
df2 = pd.read_csv(file2_url)
df3 = pd.read_csv(file3_url)
```

```python
print("DataFrame 1:")
print(df1.columns)

print("DataFrame 2:")
print(df2.columns)

print("DataFrame 3:")
print(df3.columns)
```

```python
# Renombrar columnas en df3 para que coincidan con df1 y df2
df3.rename(columns={'State': 'ST', 'Gender': 'GENDER'}, inplace=True)
```

```python
print("Columnas en df1:")
print(df1.columns.tolist())
print("\nColumnas en df2:")
print(df2.columns.tolist())
print("\nColumnas en df3:")
print(df3.columns.tolist())
```

```python
# Definir el orden estándar de columnas (usamos el orden de df1 como referencia)
orden_comun = [
    'Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
    'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
    'Policy Type', 'Vehicle Class', 'Total Claim Amount'
]

# Reorganizar las columnas de df2 y df3 según el orden común
df2 = df2[orden_comun]
df3 = df3[orden_comun]
```

```python
print("Columnas en df1:")
print(df1.columns.tolist())
print("\nColumnas en df2:")
print(df2.columns.tolist())
print("\nColumnas en df3:")
print(df3.columns.tolist())
```

```python
# Concatenar los tres DataFrames
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
combined_df.describe()
```

```python
print("Forma del DataFrame combinado:", combined_df.shape)
print("\nPrimeras 5 filas:")
print(combined_df.head())
print("\nColumnas:")
print(combined_df.columns.tolist())
```

```python
# Convertir nombres de columnas a minúsculas
combined_df.columns = combined_df.columns.str.lower()

# Revisar valores nulos
print("Valores nulos por columna:")
print(combined_df.isnull().sum())

# Revisar tipos de datos
print("\nTipos de datos:")
print(combined_df.dtypes)
```

```python
# Convertir la columna 'customer lifetime value' a tipo numérico
combined_df['customer lifetime value'] = pd.to_numeric(combined_df['customer lifetime value'], errors='coerce')
```

```python
print("Tipos de datos después de la conversión:")
print(combined_df.dtypes)
```

```python
# Contar filas donde todas las columnas importantes son nulas
columnas_importantes = [
    'customer', 'st', 'gender', 'education', 'customer lifetime value',
    'income', 'monthly premium auto', 'number of open complaints',
    'policy type', 'vehicle class', 'total claim amount'
]

filas_todas_nulas = combined_df[combined_df[columnas_importantes].isnull().all(axis=1)]
print("Número de filas con TODAS las columnas importantes nulas:", filas_todas_nulas.shape[0])
```

```python
# Contar filas que tienen AL MENOS un valor no nulo
filas_con_algun_dato = combined_df[combined_df[columnas_importantes].notnull().any(axis=1)]
print("Número de filas con AL MENOS un valor no nulo:", filas_con_algun_dato.shape[0])

# Contar filas que tienen AL MENOS un valor nulo (pero no todas)
filas_con_datos_parciales = combined_df[
    (combined_df[columnas_importantes].isnull().any(axis=1)) &
    (combined_df[columnas_importantes].notnull().any(axis=1))
]
print("Número de filas con datos parciales (algunos nulos, algunos no):", filas_con_datos_parciales.shape[0])
```

```python
print("Tipos de datos:")
print(combined_df.dtypes)
```

```python
# Copiar el DataFrame original
df_rellenado = combined_df.copy()

# Columnas numéricas (rellenar con mediana)
columnas_numericas = [
    'customer lifetime value', 'income', 'monthly premium auto', 'total claim amount'
]

for col in columnas_numericas:
    mediana = df_rellenado[col].median()
    df_rellenado[col] = df_rellenado[col].fillna(mediana)

# Columnas categóricas (rellenar con 'Desconocido')
columnas_categoricas = [
    'gender', 'education', 'policy type', 'vehicle class'
]

for col in columnas_categoricas:
    df_rellenado[col] = df_rellenado[col].fillna('Desconocido')
```

```python
print("Valores nulos después del rellenado:")
print(df_rellenado.isnull().sum())
```

```python
df_rellenado.to_csv('combined_clean_data_filled.csv', index=False)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Cargar el archivo de análisis de marketing
url_marketing = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

df_marketing = pd.read_csv(url_marketing)
```

```python
df_marketing.head()
```

```python
print("Forma del DataFrame:", df_marketing.shape)
print("\nColumnas:")
print(df_marketing.columns.tolist())
print("\nPrimeras 5 filas:")
print(df_marketing.head())
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Crear la tabla dinámica
pivot_table = df_marketing.pivot_table(
    values='total_claim_amount',  # La columna que queremos sumar
    index='sales_channel',        # La columna que agrupamos (filas)
    aggfunc='sum'                 # La función que aplicamos (sumar)
)
```

```python
# Redondear a 2 decimales
pivot_table = pivot_table.round(2)

print("Tabla de ingresos totales por canal de ventas:")
print(pivot_table)
```

La tabla muestra:

Agent: 1,810,226.82
Branch: 1,301,204.00
Call Center: 926,600.82
Web: 706,600.04
Conclusión: Los canales Agent y Branch son los que generaron más ingresos totales, siendo Agent el canal con mayores ventas.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Crear la tabla dinámica
pivot_clv = df_marketing.pivot_table(
    values='customer_lifetime_value',  # La columna que queremos promediar
    index='gender',                    # Agrupar por género en las filas
    columns='education',               # Agrupar por educación en las columnas
    aggfunc='mean'                     # Calcular el promedio
).round(2)  # Redondear a 2 decimales

print("Valor de vida promedio por género y educación:")
print(pivot_clv)
```

Insights (conclusiones):

Nivel educativo más valioso:
Para mujeres (F): High School or Below tiene el valor más alto (8675.22).
Para hombres (M): College tiene el valor más alto (8052.46).
Diferencias por género:
Las mujeres con High School or Below tienen un valor de vida muy superior al resto.
Los hombres con College tienen un valor de vida muy superior al resto.
En general, los hombres tienden a tener valores ligeramente más altos en College y Master.
Nivel educativo menos valioso:
Para ambos géneros, Doctor es el nivel con el valor de vida más bajo (mujeres: 7328.51, hombres: 7415.33).
